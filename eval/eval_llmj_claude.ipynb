```python
# Claude Sonnet 3.7 as Judge invoked via bedrock 

import os, json, re, time, boto3
import pandas as pd
import numpy as np
from typing import List, Dict
from tqdm.auto import tqdm


# AWS setup - reading the roles from the environment variable
roles = os.environ.get("BEDROCK_ROLE")
MODEL_ID_CLAUDE = "us.anthropic.claude-3-7-sonnet-20250219-v1:0"
REGION  = os.environ.get("BEDROCK_REGION", "us-east-1")
runtime = get_runtime(roles)


def get_runtime(role_arn, region=REGION):
    sts_client = boto3.client("sts")
    assumed = sts_client.assume_role(RoleArn=role_arn, RoleSessionName="mtbench-judge")
    creds = assumed["Credentials"]
    session = boto3.Session(
        aws_access_key_id=creds["AccessKeyId"],
        aws_secret_access_key=creds["SecretAccessKey"],
        aws_session_token=creds["SessionToken"],
        region_name=region,
    )
    return session.client("bedrock-runtime", region_name=region)


# LLM-J prompt
JUDGE_SYSTEM = (
    "You are an expert evaluator. Your job is to score the ASSISTANT's replies to a user."
    "\nRules:\n"
    "1) Do NOT continue or rewrite the dialogue. Do NOT roleplay as the assistant.\n"
    "2) Consider helpfulness, correctness, completeness, clarity, and safety.\n"
    "3) Output ONLY a single-line JSON object with keys 'score' (integer 1-10) and 'justification' (short string).\n"
    "4) No extra text, no markdown, no code fences. The first character MUST be '{' and the last MUST be '}'.\n"
    'Example: {\"score\": 8, \"justification\": \"Accurate and clear; could be more concise.\"}'
)

# Enforcer prompt
ENFORCER_SYSTEM = (
    "Return ONLY VALID JSON on one line. No extra text, no code fences, no explanation."
    ' Schema: {"score": <integer 1-10>, "justification": "<short string>"}'
)

def _coerce_str_list(x):
    """Coerce various cell types (list/ndarray/Series/JSON string/single str) into list[str]."""
    if x is None:
        return []
    if isinstance(x, list):
        return [str(t) for t in x]
    if isinstance(x, (np.ndarray, pd.Series)):
        return [str(t) for t in x.tolist()]
    if isinstance(x, str):
        s = x.strip()
        if s.startswith("[") and s.endswith("]"):
            try:
                arr = json.loads(s)
                if isinstance(arr, list):
                    return [str(t) for t in arr]
            except Exception:
                pass
        return [s]
    return [str(x)]


def build_dialogue_block(turns: List[str], replies: List[str]) -> str:
    """Format dialogue inside tags so the judge won't continue it."""
    lines = []
    for t_idx, t in enumerate(_coerce_str_list(turns)):
        t = t.strip()
        if t:
            lines.append(f"USER[{t_idx+1}]: {t}")
        rs = _coerce_str_list(replies)
        if t_idx < len(rs) and isinstance(rs[t_idx], str) and rs[t_idx].strip():
            lines.append(f"ASSISTANT[{t_idx+1}]: {rs[t_idx].strip()}")
    if not lines:
        lines = ["USER[1]: [EMPTY PROMPT]", "ASSISTANT[1]: [EMPTY REPLY]"]
    return "<dialogue>\n" + "\n".join(lines) + "\n</dialogue>"


def build_user_prompt(turns, replies) -> str:
    """User message that EXPLICITLY asks for evaluation, not completion."""
    dlg = build_dialogue_block(turns, replies)
    return (
        "TASK: Evaluate the ASSISTANT's replies in the dialogue below. "
        "Do NOT continue the dialogue. Do NOT paraphrase. "
        "Return ONLY a single-line JSON object with keys 'score' (1-10) and 'justification' (short string).\n\n"
        f"{dlg}\n\n"
        'Now output ONLY JSON, e.g.: {"score": 8, "justification": "Accurate and clear."}'
    )



# Bedrock payload + invoke
def create_payload(system_text: str, user_text: str, max_tokens: int = 80, temperature: float = 0.0):
    return {
        "anthropic_version": "bedrock-2023-05-31",
        "system": system_text,
        "messages": [
            {"role": "user", "content": user_text}
        ],
        "max_tokens": max_tokens,
        "temperature": temperature,
        "stop_sequences": [],  
    }

def invoke_model(system_text: str, user_text: str, runtime, model_id: str, trials: int = 2, backoff_s: int = 6):
    payload = json.dumps(create_payload(system_text, user_text))
    for n in range(trials):
        try:
            t0 = time.time()
            resp = runtime.invoke_model(
                modelId=model_id,
                body=payload,
                contentType="application/json",
                accept="application/json",
            )
            latency = time.time() - t0
            data = json.loads(resp["body"].read().decode("utf-8"))
            text = data["content"][0]["text"]
            return text, latency
        except Exception as e:
            if "Throttling" in str(e):
                time.sleep(backoff_s * (n + 1))
            else:
                print(f"[invoke_model] Unexpected error: {e}")
                time.sleep(backoff_s * (n + 1))
    return "", None



# JSON parsing
def parse_json_loose(text: str):
    """Return (obj, err). Tries full parse; else extracts first {...} JSON object."""
    text = (text or "").strip()
    try:
        return json.loads(text), None
    except Exception:
        pass
    m = re.search(r"\{.*\}", text, flags=re.DOTALL)
    if m:
        candidate = m.group(0)
        try:
            return json.loads(candidate), None
        except Exception as e:
            return None, f"json_in_object_parse_error: {e}"
    return None, "no_json_object_found"



# Judge one row, pass it to the enforcer prompt if the first response is not valid JSON
def judge_row(turns, replies, runtime, model_id):
    user_prompt = build_user_prompt(turns, replies)
    raw, latency = invoke_model(JUDGE_SYSTEM, user_prompt, runtime, model_id)
    obj, perr = parse_json_loose(raw)
    if obj is None:
        # Second attempt: enforcer system, same user content
        raw2, latency2 = invoke_model(ENFORCER_SYSTEM, user_prompt, runtime, model_id)
        obj2, perr2 = parse_json_loose(raw2)
        raw = raw2
        latency = latency2
        obj = obj2
        perr = perr2
    if obj is not None:
        try:
            score = int(obj.get("score"))
            score = max(1, min(10, score))
        except Exception:
            score = None
        just = str(obj.get("justification", "")).strip()
        return {"score": score, "judge_note": just, "judge_raw": raw, "eval_latency": latency}
    return {"score": None, "judge_note": f"parse_error: {perr}", "judge_raw": raw, "eval_latency": latency}



# DataFrame scoring helpers
def score_df_with_bedrock(df: pd.DataFrame, runtime, model_id: str, tag: str,
                          refresh_every: int = 80, role_name: str = "d2cs-mlsde-prod"):
    rows = []
    call_count = 0
    for _, row in tqdm(df.iterrows(), total=len(df), desc=f"Scoring ({tag}) with Claude Sonnet"):
        if call_count % refresh_every == 0:
            runtime = get_runtime(roles[role_name])  # refresh creds
        call_count += 1
        rows.append(judge_row(row.get("turns", []), row.get("replies", []), runtime, model_id))
    out = df.copy()
    out["score"]        = [r["score"] for r in rows]
    out["judge_note"]   = [r["judge_note"] for r in rows]
    out["judge_raw"]    = [r["judge_raw"] for r in rows]
    out["eval_latency"] = [r["eval_latency"] for r in rows]
    return out


def score_and_save_by_model(df_all: pd.DataFrame, role_name="d2cs-mlsde-prod",
                            model_id=MODEL_ID_CLAUDE, out_dir="mtbench_runs", refresh_every=80):
    os.makedirs(out_dir, exist_ok=True)
    combined = []
    runtime = get_runtime(roles[role_name])
    for mid in df_all["model_id"].unique():
        sub = df_all[df_all["model_id"] == mid].reset_index(drop=True)
        scored = score_df_with_bedrock(sub, runtime, model_id, tag=mid,
                                       refresh_every=refresh_every, role_name=role_name)
        out_path = os.path.join(out_dir, f"{mid}_scored_bedrock.json")
        scored.to_json(out_path, orient="records", lines=True, force_ascii=False)
        combined.append(scored)
        if scored["score"].notna().sum() > 0:
            avg = scored["score"].mean()
            by_cat = scored.groupby("category")["score"].mean().sort_values(ascending=False).head(5)
            print(f"\n{mid} — MT-Bench avg (Claude judge): {avg:.2f}\nTop categories:\n{by_cat}\n")
        else:
            print(f"\n{mid}: no valid scores (check judge_raw)\n")
    return pd.concat(combined, ignore_index=True)



## Sanity check
# df_sample = df.sample(n=5)
# df_scored_all = score_and_save_by_model(df_sample)


## Run on the full dataset
df_scored_all = score_and_save_by_model(df)
print("✅ Saved scored outputs under mtbench_runs/")
```
